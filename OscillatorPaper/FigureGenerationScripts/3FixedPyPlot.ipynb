```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize
# import seaborn as sns
```

```python
# Read the CSV file
file_path = '/Users/jonathanfischer/Desktop/PhD_ThesisWork/Julia/OscillatorPaper/FigureGenerationScripts/fixed_triplet_results-kcat1kcat7DF.csv'
data = pd.read_csv(file_path)
```

```python
# Clean the data by removing extreme values
data = data.replace([np.inf, -np.inf], np.nan).dropna()
data = data[(data != 0).all(1)]
```

```python
data
```

```python
# Log scale the parameters
log_kcat1_cleaned = np.log10(data['kcat1'])
log_kcat7_cleaned = np.log10(data['kcat7'])
log_DF_cleaned = np.log10(data['DF'])
```

```python
# Normalize the sizes for the scatter plot
sizes = Normalize()(data['average_amplitude'])

# Define a colormap for the plots
cmap = 'viridis'

# Define angles for different views
angles = [0, 30, 60, 90, 120, 150]
```

```python
# Set a colormap
cmap = plt.get_cmap('viridis')

# Reload the dataset
fixed_triplet_results_kcat1kcat7DF = data

# Define a threshold to identify extreme values (values close to zero)
threshold = 1e-9

# Replace extreme values with 0 in the relevant columns
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['kcat1'] < threshold, 'kcat1'] = 0
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['kcat7'] < threshold, 'kcat7'] = 0
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['DF'] < threshold, 'DF'] = 0

# Define the log-scaled parameters (with epsilon added to avoid taking log of zero)
epsilon = 1e-9
log_kcat1_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['kcat1'] + epsilon)
log_kcat7_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['kcat7'] + epsilon)
log_DF_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['DF'] + epsilon)

# Normalize the average amplitude and period for the point size and color
fixed_triplet_results_kcat1kcat7DF['average_amp_norm'] = (fixed_triplet_results_kcat1kcat7DF['average_amplitude'] - fixed_triplet_results_kcat1kcat7DF['average_amplitude'].min()) / (fixed_triplet_results_kcat1kcat7DF['average_amplitude'].max() - fixed_triplet_results_kcat1kcat7DF['average_amplitude'].min())
fixed_triplet_results_kcat1kcat7DF['average_per_norm'] = (fixed_triplet_results_kcat1kcat7DF['average_period'] - fixed_triplet_results_kcat1kcat7DF['average_period'].min()) / (fixed_triplet_results_kcat1kcat7DF['average_period'].max() - fixed_triplet_results_kcat1kcat7DF['average_period'].min())

# Create 3D scatter plot with projections onto each plane
fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(111, projection='3d')

# Create the main 3D scatter plot
sc = ax.scatter(log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned,
                c=fixed_triplet_results_kcat1kcat7DF['average_per_norm'],
                s=fixed_triplet_results_kcat1kcat7DF['average_amp_norm']*500,
                alpha=0.6, cmap=cmap, marker='o')

# Project the points onto each plane and connect them with lines
for x, y, z in zip(log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned):
    ax.plot([x, x], [y, y], [z, log_DF_cleaned.min()], c='gray', alpha=0.2) # Projection onto XY plane
    ax.plot([x, x], [y, log_kcat7_cleaned.max()], [z, z], c='gray', alpha=0.2) # Projection onto XZ plane
    ax.plot([x, log_kcat1_cleaned.max()], [y, y], [z, z], c='gray', alpha=0.2) # Projection onto YZ plane

# Set larger axis labels
ax.set_xlabel('log10(kcat1)', fontsize=14)
ax.set_ylabel('log10(kcat7)', fontsize=14)
ax.set_zlabel('log10(DF)', fontsize=14)

# Add a colorbar
plt.colorbar(sc, ax=ax, label='Average Period')

# Set the title
ax.set_title('Log-Scaled Parameters with Projections and Lines (Cleaned Data) - Size: Average Amplitude, Color: Average Period', fontsize=14)

plt.show()
```

```python
# Function to create 3D scatter plot with accurate color-matching projections (shadows) on all walls
def create_3d_scatter_with_accurate_shadows(angle, x, y, z, sizes, colors):
    fig = plt.figure(figsize=(14, 7), dpi=200, )
    ax = fig.add_subplot(111, projection='3d')

    ax.set_xlim([-2, 3])
    ax.set_ylim([-2, 3])
    ax.set_zlim([3, 5])

    # Normalize the colors for accurate mapping
    norm_colors = Normalize()(colors)

    # Create the main 3D scatter plot with reduced alpha
    sc = ax.scatter(x, y, z, c=colors, s=sizes * 500, alpha=0.3, cmap=cmap, marker='o')

    # Get the colormap to extract RGBA values
    colormap = plt.cm.get_cmap(cmap)

    # Project the points onto each plane and connect them with lines, including color-matching shadows
    for xi, yi, zi, si, ci in zip(x, y, z, sizes, norm_colors):
        rgba_color = colormap(ci)
        ax.plot([xi, xi], [yi, y.min()], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(xi, y.min(), zi, color=rgba_color, s=si * 500, alpha=0.4, marker='o')
        ax.plot([xi, x.min()], [yi, yi], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(x.min(), yi, zi, color=rgba_color, s=si * 500, alpha=0.4, marker='o')
        ax.plot([xi, xi], [yi, yi], [z.min(), zi], c='gray', alpha=0.2)
        ax.scatter(xi, yi, z.min(), color=rgba_color, s=si * 500, alpha=0.4, marker='o')

    ax.set_xlabel('log10(kcat1)', fontsize=14)
    ax.set_ylabel('log10(kcat7)', fontsize=14)
    ax.set_zlabel('log10(DF)', fontsize=14)
    ax.view_init(elev=10, azim=angle)
    plt.colorbar(sc, ax=ax, label='Average Period')
    ax.set_title(f'Log-Scaled Parameters with Accurate Color-Matching Projections & Shadows (Angle: {angle}°) - Size & Color: Average Amplitude & Period', fontsize=14)
    plt.show()

# Plot from different angles with log scaling and accurate color-matching shadows
for angle in angles:
    create_3d_scatter_with_accurate_shadows(angle, log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data['average_period'])
```
